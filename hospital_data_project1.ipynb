### QUESTION 3
To make a staffing plan, which decides the number of nurses and doctors needed to serve patients, a deep understanding of patient arrivals is crucial. The staffing plan in practice will be made on an hourly basis (24 intervals) every day. Thus, please create a new column, REGIS_HOUR in df. Moreover, the patients’ arrival pattern may vary by the day of a month. Please also create a new column, REGIS_DAY in df.

```python
df['REGIS_HOUR'] = pd.to_datetime(df.REGIS_DATE + ' ' + df.REGIS_TIME).dt.hour
df['REGIS_DAY'] = pd.to_datetime(df.REGIS_DATE + ' ' + df.REGIS_TIME).dt.day
df['regis'] = pd.to_datetime(df.REGIS_DATE + ' ' + df.REGIS_TIME)
df
```

### QUESTION 4
Find out the average number of patient arrivals in each hour of a day. To answer this question, we assume the arrival pattern is similar across different days and only utilise the data in 2011.

```python
average_patients_per_hour = pd.read_csv('EDdata_new.csv', parse_dates=[[2, 3]])
#pd.read_csv?
```

```python
average_patients_per_hour['REGIS_DATE'] = average_patients_per_hour.REGIS_DATE_REGIS_TIME.dt.date
average_patients_per_hour['REGIS_HOUR'] = average_patients_per_hour.REGIS_DATE_REGIS_TIME.dt.hour
```

```python
average_patients_per_hour[['Case', 'REGIS_DATE', 'REGIS_HOUR', 'REGIS_DATE_REGIS_TIME']][average_patients_per_hour.REGIS_DATE_REGIS_TIME <'2012-01-01'].groupby(['REGIS_HOUR'])['REGIS_HOUR'].count()
```

```python
average_patients_per_hour[['Case', 'REGIS_HOUR', 'REGIS_DATE_REGIS_TIME']][average_patients_per_hour.REGIS_DATE_REGIS_TIME <'2012-01-01'].groupby(['REGIS_HOUR'])['REGIS_HOUR'].count() / average_patients_per_hour[average_patients_per_hour.REGIS_DATE_REGIS_TIME < '2012-01-01'].REGIS_DATE.nunique()
```

The assumption that the arrival pattern is similar across different days is too strong to be true. Let's discuss the weekday effect (including Saturday and Sunday) on the arrival pattern of patients. ### QUESTION 6
Using the WEEKDAY column, please find out the average number of patient arrivals in each hour by weekday categories. Your answer should be a 7-by-24 table.

```python
pivoted = (average_patients_per_hour.query('REGIS_DATE_REGIS_TIME < "2012-01-01"')
           .pivot_table(
               'Case',
               index='REGIS_HOUR',
               columns='REGIS_Weekday',
               aggfunc='count')
           )
pivoted
```
